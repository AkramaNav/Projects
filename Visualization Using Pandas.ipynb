# Pandas Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Akrama\Desktop\Importing Files in Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('Solarize_Light2')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 250, c = 'Green')
```

```python
df.plot.hist(bins  = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating' , figsize = (10,10))
```
