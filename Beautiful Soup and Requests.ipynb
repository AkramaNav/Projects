# Beautiful Soup and Requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```
